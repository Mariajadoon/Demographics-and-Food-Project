```python
import pandas as pd
import numpy as np

file = "acs2015_census_tract_data.csv"
df = pd.read_csv(file, encoding="ISO-8859-1")
df.groupby("State")
df.head()
```

```python
clean_data = df[["State", "TotalPop","Men","Women","Hispanic","White","Black","Native",
          "Asian","Pacific","Income","IncomePerCap","Poverty",
         "Professional","Service","Office","Construction","Production",
         "Employed","PrivateWork","PublicWork","SelfEmployed","FamilyWork","Unemployment"]]
clean_data.head()
```

```python
clean_data.groupby("State")
clean_data.head()
```

```python
population = clean_data.groupby(["State"]).sum()["TotalPop"]
men = clean_data.groupby(["State"]).sum()["Men"]
women = clean_data.groupby(["State"]).sum()["Women"]
hispanic = clean_data.groupby(["State"]).mean()["Hispanic"]
white = clean_data.groupby(["State"]).mean()["White"]
black = clean_data.groupby(["State"]).mean()["Black"]
native = clean_data.groupby(["State"]).mean()["Native"]
asian = clean_data.groupby(["State"]).mean()["Asian"]
pacific = clean_data.groupby(["State"]).mean()["Pacific"]
income = clean_data.groupby(["State"]).mean()["Income"]
incomepercap = clean_data.groupby(["State"]).mean()["IncomePerCap"]
poverty = clean_data.groupby(["State"]).mean()["Poverty"]
professional = clean_data.groupby(["State"]).mean()["Professional"]
service = clean_data.groupby(["State"]).mean()["Service"]
office = clean_data.groupby(["State"]).mean()["Office"]
construction = clean_data.groupby(["State"]).mean()["Construction"]
production = clean_data.groupby(["State"]).mean()["Production"]
employed = clean_data.groupby(["State"]).mean()["Employed"]
privatework = clean_data.groupby(["State"]).mean()["PrivateWork"]
publicwork = clean_data.groupby(["State"]).mean()["PublicWork"]
selfemployed = clean_data.groupby(["State"]).mean()["SelfEmployed"]
familywork = clean_data.groupby(["State"]).mean()["FamilyWork"]
unemployment = clean_data.groupby(["State"]).mean()["Unemployment"]

state_summary = pd.DataFrame({"TotalPop":population,
                              "Men":men,
                              "Women":women,
                              "Hispanic":hispanic,
                              "White":white,
                              "Black":black,
                              "Native":native,
                              "Asian":asian,
                              "Pacific":pacific,
                              "Income":income,
                              "IncomePerCap":incomepercap,
                              "Poverty":poverty,
                              "Professional":professional,
                              "Service":service,
                              "Office":office,
                              "Construction":construction,
                              "Production":production,
                              "Employed":employed,
                              "PrivateWork":privatework,
                              "PublicWork":publicwork,
                              "SelfEmployed":selfemployed,
                              "FamilyWork":familywork,
                              "Unemployment":unemployment})
state_summary.head()
```

```python
state_summary.to_csv("project_data.csv",
                  encoding="utf-8", index=False, header=True)
```
